```python
#perform historical volatility and risk-returns ratios
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#compute the log_returns
log_returns = np.log(df.Close/df.Close.shift(1)).dropna()
print(log_returns)

#visualize the returns
#returns should be high and stable as possible
log_returns.plot().update_layout(autosize=False, width=900, height=600)

```

```python
#calculate the daily standard deviation checks volatility
#higher volatility ..more money ...higher risks
daily_std = log_returns.std()
print(daily_std)

print("\n\n\n")
#calculates the annual volatility of working days
annualized_vol = daily_std*np.sqrt(252)
annualized_vol*100
```

```python
#plot the histogram for log_returns with annualized volatility
fig = make_subplots(rows=2, cols=2)#for the four stock markets
trace0 = go.Histogram(x=log_returns['AAPL'], name = 'Apple Inc')
trace1 = go.Histogram(x=log_returns['MSFT'], name = 'Microsoft Inc')
trace2 = go.Histogram(x=log_returns['GOOG'], name = 'Google Corp')
trace3 = go.Histogram(x=log_returns['GME'], name = 'GameStop Inc')

#append the trace to the fig
fig.append_trace(trace0, 1,1)
fig.append_trace(trace1, 1,2)
fig.append_trace(trace2, 2,1)
fig.append_trace(trace3, 2,2)

fig.update_layout(autosize=False, width=600, height=600, title="Frequency of log_returns",
                 xaxis=dict(title='AAPL Annualized Vol: '+str(np.round(annualized_vol['AAPL']*100,1))),
                 xaxis2=dict(title='MSFT Annualized Vol: '+str(np.round(annualized_vol['MSFT']*100,1))),
                 xaxis3=dict(title='GOOG Annualized Vol: '+str(np.round(annualized_vol['GOOG']*100,1))),
                 xaxis4=dict(title='GME Annualized Vol: '+str(np.round(annualized_vol['GME']*100,1))),
                 )


fig.show()

```

```python
#trailing volatility over time
#gives the markets trends either uptrends or downtrends of the market
Trading_Days=60 #trading window
volatility = log_returns.rolling(window=Trading_Days).std()*np.sqrt(Trading_Days)
volatility.plot().update_layout(autosize=False, width=900, height=600)
#buy during the downturn and sell during the upturn
```

```python
#calculate the risks
#we implement the sharpe ratio, which returns the risk adjusted return
#set a risk free rate per unit volatility
rf = 0.01/252 #training days in a year
sharpe_ratio = (log_returns.rolling(window=Trading_Days).mean() - rf)*Trading_Days/volatility
```

```python
#Sharpe ratio greater than 1.0 is considered acceptable to good by investors.
#A ratio higher than 2.0 is rated as very good.
#A ratio of 3.0 or higher is considered excellent. 
#A ratio under 1.0 is considered sub-optimal.
#plot the sharpe_ration graph
sharpe_ratio.plot().update_layout(autosize=False, width=900, height=600)
```

```python
#machine learning model, SUPVERSIVED PROBLEM 
#under regression .....forecasts future stock prices 
#implements sklearn model API(fbprophet)
#call the instance of prophet class and call its fits and predict methods
#prophet two inputs the ds and y, ds rep the datestamp column and the y rep the mesurement we want to forecast.
```

```python
#import the Prophet class
#Prophet uses a linear model for its forecast.
from prophet import Prophet
```

```python
#set the two main inputs ds(datestamp) and y (the measurements to forecast)
#we will forecast the Close price of the stock markets
data = pdr.get_data_yahoo('AAPL', start, end)
data.reset_index(inplace=True)
df_train = data[['Date','Close']]
df_train = df_train.rename(columns = {'Date':"ds", "Close":"y"})
df_train.head()
```

```python
#fit the model
model = Prophet()
model.fit(df_train)
#forecast the future on the ds (date column) given a no of days
future = model.make_future_dataframe(periods=65)
future.tail()
```

```python
#predictions
#predict methods assigns each row in future a predicted value which it names yhat
forecast = model.predict(future)
print("<..................Predicted Prices....................>")
#to predict on specified prices 
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())


print("\n\n\n")
#displays the predicted future prices on daily, weekly, monthly and yearly basis
print(forecast.tail())
```

```python
#plot the forecast by calling the Prophet.plot method 
#pass in your forecast dataframe.
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(model, forecast)
```

```python
#check automatic trend changes
#by default the changepoints are set to 25
from prophet.plot import add_changepoints_to_plot
fig = model.plot(forecast)
#get the current axis
a = add_changepoints_to_plot(fig.gca(), model, forecast)
```

```python
#compare how the model performs to actual values
#use R-Squared and Mean Squared Error.
#combined dataframe with yhat from the forcasts and the original y values from the data
metric_df = forecast.set_index('ds')[['yhat']].join(df_train.set_index('ds').y).reset_index()
metric_df.tail()
```

```python
#drop the NaN values
metric_df.dropna(inplace=True)
```

```python
metric_df.tail()
#with yhat rep the forecast data and y rep the actual data
```

```python
#import the metrics modeules
from sklearn.metrics import r2_score
```

```python
#calculate the R_Squared value
#the model has a 99% accuracy
r2_score(metric_df.y, metric_df.yhat)
```
